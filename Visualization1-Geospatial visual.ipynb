```python
import plotly.express as px
for day in range(0,470,7):
    death_count = list(df.iloc[day])[3:]
    state_data['Deaths'] = death_count
    state_data['Deaths'] = state_data['Deaths'].fillna(0)
    state_data['Deaths'] = state_data['Deaths'].replace(-1,0)
    state_data['Deaths'] = state_data['Deaths']*6

    fig = px.scatter_mapbox(state_data, lat="Latitude", lon="Longitude", size='Deaths',size_max=80,
                            color_discrete_sequence=["red"],opacity = 0.5, zoom=4.3, width = 1000,height=900, title=f'Date:{df.iloc[day][0]}')
    fig.update_layout(title_x = 0.5, title_y = 0.98,mapbox_style="carto-darkmatter") #stamen-toner
    fig.update_layout(margin={"r":10,"t":50,"l":0,"b":0,"l":10})
    total_deaths = sum(df['TT'][:day])
    fig.add_annotation(
            x=0.8,
            y=0.2,
            text= f'Total Deaths: {total_deaths}',
            showarrow=False,
            font=dict(
                family="Courier New, monospace",
                size=20,
                color="#ffffff"
                ),
            bordercolor="#c7c7c7",
            borderwidth=2,
            borderpad=4,
            bgcolor="red",
            opacity=0.8
            )#     fig.update_layout(title=f'Date:{df.iloc[day][0]}')
    fig.write_image(f'newer_f/file_{day}.png')
#     fig.show()
```
